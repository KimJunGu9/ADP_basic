```python
# 원본 데이터에서 범주형 변수를 대체하기
```

```python
df_dum = pd.get_dummies(df,columns=['주구매상품','주구매지점'])
```

```python
df_dum
```

```python
# dummies를 사용하지않는다면 아래와 같이 하면 된다.
# 그러나 컬럼명이 어느것이 주구매상품인지, 어느것이 주구매지점인지 알 수 없으므로 위의 방법으로 사용하자

pd.concat([df,pd.get_dummies(df['주구매상품']),pd.get_dummies(df['주구매지점'])],axis=1).drop(columns=['주구매상품','주구매지점'])
```
