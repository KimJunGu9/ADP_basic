```python
# column 기준 정렬

# df.sort_index(axis=1) : df 객체 column명으로 정렬
# default : 오름차순
# ascending = False : 내림차순
```

```python
df_t.sort_index(axis=1, ascending = False)
```

```python
# 값 기준 정렬

# df.sort_values() : 시리즈 객체 값 기준으로 오름차순 정렬
# df.sort_values(by='컬럼명) : df 객체를 지정한 컬럼의 값 기준으로 정렬
# df.sort_values(by=['컬럼명1','컬럼명2']) : df 객체를 지정한 컬럼들의 값 기준으로 정렬
```

```python
df_t.sort_values(by=['age','fare'], ascending = False)
```

```python
df_t.sort_values(by = ['survived','age'])
```
